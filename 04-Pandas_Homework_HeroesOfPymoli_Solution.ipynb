```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
heroes_of_pymoli_csv = "04-Pandas_Homework_HeroesOfPymoli_Resources_purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(heroes_of_pymoli_csv)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Gather unique "SN" values.
total_players = [{"Total Players": len(purchase_data["SN"].unique())}]

# Format into Data Frame
total_players_df = pd.DataFrame(total_players)

total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Gather unique "Item Name" values.
unique_items = len(purchase_data["Item Name"].unique())

# Find average item price
average_price = round(purchase_data["Price"].mean(), 2)

# Find total purchases
total_purchases = len(purchase_data.index)

# Sum together "Price" column to find total revenue.
total_revenue = purchase_data["Price"].sum()

# Store values in dictionary
summary_dict = [{"Number of Unique Items": unique_items, "Average Price": average_price, "Number of Purchases": total_purchases, "Total Revenue": total_revenue}]

# Convert dictionary to Data Frame
summary_df = pd.DataFrame(summary_dict)

summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Drop duplicate "SN" data
purchase_data_clean = purchase_data.drop_duplicates(subset = ["SN"])

# Find total number of each gender
player_genders_count = purchase_data_clean["Gender"].value_counts()

# Convert to Data Frame
player_genders_count_df = pd.DataFrame(player_genders_count)

# Find percentage values of each gender
player_genders_percent = round((purchase_data_clean["Gender"].value_counts("Male"))*100, 2)

# Convert to Data Frame
player_genders_percent_df = pd.DataFrame(player_genders_percent)

# Concatenate Data Frames together
player_genders_df = pd.concat([player_genders_count_df, player_genders_percent_df], axis=1)

# Rename columns
player_genders_df.columns = ["Total Count", "Percentage of Players"]

player_genders_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Set index to "Gender"
purchase_genders = purchase_data.set_index("Gender")

# Find total purchase count for each gender
purchase_count = purchase_data["Gender"].value_counts()

# Convert to Data Frame
purchase_count_df = pd.DataFrame(purchase_count)

# Find average purchase price for each gender
average_purchase_genders = round(purchase_genders.groupby(level=0).mean(), 2)

# Delete unecessary columns
del average_purchase_genders["Purchase ID"]

del average_purchase_genders["Item ID"]

del average_purchase_genders["Age"]

# Find the total purchase value for each gender
total_purchase_genders = purchase_genders.groupby(level=0).sum()

# Delete unecessary columns
del total_purchase_genders["Purchase ID"]

del total_purchase_genders["Item ID"]

del total_purchase_genders["Age"]

# Find the total purchase value per person
purchase_genders_person = purchase_genders.groupby(["SN"])["Price"].transform("sum")

# Convert to Data Frame
purchase_genders_person_df = pd.DataFrame(purchase_genders_person)

# Concatenate with gender index Data Frame
purchase_genders_total = pd.concat([purchase_genders, purchase_genders_person_df], axis=1)

# Drop duplicate "SN" values
purchase_genders_total_clean = purchase_genders_total.drop_duplicates(subset = ["SN"])

# Rename columns so there are no longer two "Price" columns 
purchase_genders_total_clean.columns = ["Purchase ID", "SN", "Age", "Item ID", "Item Name", "Price", "Total Money Spent"]

# Delete unecessary columns
del purchase_genders_total_clean["Item ID"]

del purchase_genders_total_clean["Item Name"]

del purchase_genders_total_clean["Price"]

del purchase_genders_total_clean["Purchase ID"]

del purchase_genders_total_clean["Age"]

# Find the average total purchase price per person for each gender
average_purchase_genders_total = round(purchase_genders_total_clean.groupby(level=0).mean(), 2)

# Concatenate the Data Frames
summary_gender_purchases = pd.concat([purchase_count_df, average_purchase_genders, total_purchase_genders, average_purchase_genders_total], axis=1, sort=False)

# Rename columns
summary_gender_purchases.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]

summary_gender_purchases
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]

# Establish group names
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorize the existing players using the age bins
purchase_data_clean["Age Range"] = pd.cut(purchase_data_clean["Age"], bins, labels=group_names)

# Find the total count for each age group
age_group_count = purchase_data_clean["Age Range"].value_counts()

# Convert to Data Frame
age_group_count_df = pd.DataFrame(age_group_count)

# Find the percentage of each age group
age_group_percentages = round((purchase_data_clean["Age Range"].value_counts("0 to 10"))*100, 2)

# Convert to Data Frame
age_group_percentages_df = pd.DataFrame(age_group_percentages)

# Concatenate the Data Frames
summary_age_df = pd.concat([age_group_count_df, age_group_percentages_df], axis=1)

# Rename columns
summary_age_df.columns = ["Total Count", "Percentage of Players"]

# Set the age range as the index
summary_age_df = summary_age_df.reindex(index = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"])

summary_age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Bin the purchase_data Data Frame by age
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_names)

# Set the age range as the index
purchase_age = purchase_data.set_index("Age Range")

# Find the purchase count for each age range
purchase_count_age = purchase_data["Age Range"].value_counts()

# Convert to Data Frame
purchase_count_age_df = pd.DataFrame(purchase_count_age)

# Find the average purchase price for each age range
average_purchase_age = round(purchase_age.groupby(level=0).mean(), 2)

# Delete unecessary columns
del average_purchase_age["Item ID"]

del average_purchase_age["Age"]

del average_purchase_age["Purchase ID"]

# Find the total purchase value for each age range
total_purchase_age = purchase_age.groupby(level=0).sum()

# Delete unecessary columns
del total_purchase_age["Item ID"]

del total_purchase_age["Age"]

del total_purchase_age["Purchase ID"]

# Find the total purchase value per person
purchase_age_person = purchase_age.groupby(["SN"])["Price"].transform("sum")

# Convert to Data Frame
purchase_age_person_df = pd.DataFrame(purchase_age_person)

# Concatenate with age range index Data Frame
purchase_age_total = pd.concat([purchase_age, purchase_age_person_df], axis=1)

# Drop duplicate "SN" values
purchase_age_total_clean = purchase_age_total.drop_duplicates(subset = ["SN"])

# Rename columns so there are no longer two "Price" columns
purchase_age_total_clean.columns = ["Purchase ID", "SN", "Age", "Gender", "Item ID", "Item Name", "Price", "Total Money Spent"]

# Delete unecessary columns
del purchase_age_total_clean["Item ID"]

del purchase_age_total_clean["Item Name"]

del purchase_age_total_clean["Price"]

del purchase_age_total_clean["Purchase ID"]

del purchase_age_total_clean["Gender"]

del purchase_age_total_clean["Age"]

# Find the average total purchase price per person for each age range
average_purchase_age_total = round(purchase_age_total_clean.groupby(level=0).mean(), 2)

# Concatenate the Data Frames
summary_age_purchases = pd.concat([purchase_count_age_df, average_purchase_age, total_purchase_age, average_purchase_age_total], axis=1, sort=False)

# Rename columns
summary_age_purchases.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]

summary_age_purchases
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Set the "SN" column as the index
purchase_sn = purchase_data.set_index("SN")

# Find the purchase count for each SN
top_purchase_count = purchase_data["SN"].value_counts()

# Convert to Data Frame
top_purchase_count_df = pd.DataFrame(top_purchase_count)

# Find the average purchase price for each SN
top_average_purchase_price = round(purchase_sn.groupby(level=0).mean(), 2)

# Find the total purchase price for each SN
top_total_purchase_price = purchase_sn.groupby(level=0).sum()

# Concatenate Data Frames
top_spenders = pd.concat([top_purchase_count_df, top_average_purchase_price, top_total_purchase_price], axis=1, sort=False)

# Delete unecessary columns
del top_spenders["Purchase ID"]

del top_spenders["Age"]

del top_spenders["Item ID"]

# Rename columns
top_spenders.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value"]

# Sort the total purchase value column in descending order
top_spenders = top_spenders.sort_values("Total Purchase Value", ascending=False)

top_spenders.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve the Item ID, Item Name, and Item Price columns
most_popular_item_index = purchase_data[["Item ID", "Item Name", "Price"]]

# Group by Item ID and Item Name
most_popular_item_index = most_popular_item_index.groupby(["Item ID", "Item Name"])

# Find the purchase count for each item
purchase_count = most_popular_item_index["Price"].count()

# Find the total purchase value for each item
total_purchase_value = most_popular_item_index["Price"].sum()

# Find the price for each item
item_price = total_purchase_value/purchase_count

# Create a summary data frame to hold the results
most_popular_item = pd.DataFrame({"Purchase Count": purchase_count,
                                 "Item Price": item_price,
                                 "Total Purchase Value": total_purchase_value}) 

# Sort the purchase count column in descending order
most_popular_item = most_popular_item.sort_values(by=["Purchase Count"], ascending=False).head(5)

most_popular_item
```
